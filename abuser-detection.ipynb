# 4. Transformer for Bot-detection

### 캐릭터 시계열 스탯 정보들을 이용한 트랜스포머 모형 정의

```python
class MultiHeadSelfAttention(layers.Layer):
    def __init__(self, embed_dim, num_heads=8):
        super(MultiHeadSelfAttention, self).__init__()
        self.embed_dim = embed_dim
        self.num_heads = num_heads
        if embed_dim % num_heads != 0:
            raise ValueError(
                f"embedding dimension = {embed_dim} should be divisible by number of heads = {num_heads}"
            )
        self.projection_dim = embed_dim // num_heads
        self.query_dense = layers.Dense(embed_dim)
        self.key_dense = layers.Dense(embed_dim)
        self.value_dense = layers.Dense(embed_dim)
        self.combine_heads = layers.Dense(embed_dim)

    def attention(self, query, key, value):
        score = tf.matmul(query, key, transpose_b=True)
        dim_key = tf.cast(tf.shape(key)[-1], tf.float32)
        scaled_score = score / tf.math.sqrt(dim_key)
        weights = tf.nn.softmax(scaled_score, axis=-1)
        output = tf.matmul(weights, value)
        return output, weights

    def separate_heads(self, x, batch_size):
        x = tf.reshape(x, (batch_size, -1, self.num_heads, self.projection_dim))
        return tf.transpose(x, perm=[0, 2, 1, 3])

    def call(self, inputs):
        batch_size = tf.shape(inputs)[0]
        query = self.query_dense(inputs)  
        key = self.key_dense(inputs)
        value = self.value_dense(inputs)
        query = self.separate_heads(query, batch_size)
        key = self.separate_heads(key, batch_size)  
        value = self.separate_heads(value, batch_size)
        attention, weights = self.attention(query, key, value)
        attention = tf.transpose(attention, perm=[0, 2, 1, 3])
        concat_attention = tf.reshape(attention, (batch_size, -1, self.embed_dim))
        output = self.combine_heads(concat_attention)  
        return output
    
    
class TransformerBlock(layers.Layer):
    def __init__(self, embed_dim, num_heads, ff_dim, rate=0.5):
        super(TransformerBlock, self).__init__()
        self.att = MultiHeadSelfAttention(embed_dim, num_heads)
        self.ffn = keras.Sequential([
            layers.Dense(ff_dim, activation="relu"),
            layers.Dense(embed_dim),
        ])
        self.layernorm1 = layers.LayerNormalization(epsilon=1e-6)
        self.layernorm2 = layers.LayerNormalization(epsilon=1e-6)
        self.dropout1 = layers.Dropout(rate)
        self.dropout2 = layers.Dropout(rate)

    def call(self, inputs, training):
        attn_output = self.att(inputs)
        attn_output = self.dropout1(attn_output, training=training)
        out1 = self.layernorm1(inputs + attn_output)
        ffn_output = self.ffn(out1)
        ffn_output = self.dropout2(ffn_output, training=training)
        return self.layernorm2(out1 + ffn_output)
    
    
class PositionEmbedding(layers.Layer):
    def __init__(self, maxlen, embed_dim):
        super(PositionEmbedding, self).__init__()
        self.maxlen = maxlen
        self.emb = layers.Dense(embed_dim, activation='relu')
        self.pos_emb = layers.Embedding(input_dim=self.maxlen, output_dim=embed_dim)

    def call(self, x):
        positions = tf.range(start=0, limit=self.maxlen, delta=1)
        positions = self.pos_emb(positions)
        x = self.emb(x)
        return x + positions
```

# 5. 모형 학습

### 트랜스포머 모형 생성

```python
embed_dim = 32  
num_heads = 4   
ff_dim = 32     

inputs = Input(shape=(train_X.shape[1], train_X.shape[2]), dtype='float32')
embedding_layer = PositionEmbedding(maxlen, embed_dim)
x = embedding_layer(inputs)
transformer_block = TransformerBlock(embed_dim, num_heads, ff_dim)

x = transformer_block(x)
x = layers.GlobalAveragePooling1D()(x)

x = layers.Dense(16)(x)
x = layers.BatchNormalization()(x)
x = layers.Activation('relu')(x)
x = layers.Dropout(0.5)(x)

x = layers.Dense(4)(x)
x = layers.BatchNormalization()(x)
x = layers.Activation('relu')(x)
x = layers.Dropout(0.5)(x)

x = layers.Dense(1)(x)
outputs = layers.Activation('sigmoid')(x)

model = keras.Model(inputs=inputs, outputs=outputs)
```

### 모델 체크포인트, 얼리스탑핑 정의

```python
checkpoint = ModelCheckpoint('transformer.h5', monitor='val_precision', save_best_only=True, verbose=1, mode='max', save_weights_only=True)
early_stopping = EarlyStopping(monitor='val_precision', patience=200, mode='max')
```

### 분류 정량 평가 함수 정의
- recall
- precision
- f-1 score
- specificity & fallout

```python
def recall(y_target, y_pred):
    y_target = K.round(K.clip(y_target, 0, 1))
    y_pred = K.round(K.clip(y_pred, 0, 1))
    TP = K.sum(y_target * y_pred) 
    TPFN = K.sum(y_target)
    recall = TP / (TPFN + K.epsilon())
    return recall


def precision(y_target, y_pred):
    y_pred = K.round(K.clip(y_pred, 0, 1))
    y_target = K.round(K.clip(y_target, 0, 1))
    TP = K.sum(y_target * y_pred) 
    TPFP = K.sum(y_pred)
    precision = TP / (TPFP + K.epsilon())
    return precision


def fallout(y_target, y_pred):
    return 1.0 - specificity(y_target, y_pred)


def specificity(y_target, y_pred):
    y_pred = K.round(K.clip(y_pred, 0, 1))
    y_target = K.round(K.clip(y_target, 0, 1))
    neg_y_target = 1 - y_target
    neg_y_pred = 1 - y_pred
    FP = K.sum(neg_y_target * y_pred)
    TN = K.sum(neg_y_target * neg_y_pred)
    specificity = TN / (TN + FP + K.epsilon())
    return specificity


def f1(y_target, y_pred):
    _recall = recall(y_target, y_pred)
    _precision = precision(y_target, y_pred)
    _f1score = ( 2 * _recall * _precision) / (_recall + _precision+ K.epsilon())
    return _f1score
```

### 모형 학습

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy', precision, recall, f1, fallout, 'AUC'])
history = model.fit(train_X, train_y, epochs = 10000, batch_size = 512, validation_split=0.2, verbose=2, callbacks=[checkpoint, early_stopping])
```

### 캐릭터의 임베딩 벡터 및 confusion label t-SNE 시각화
t-SNE으로 2차원 임베딩된 벡터와 confusion 라벨 시각화를 통해 휴먼은 휴먼으로 잘 분류 하는 것을 확인 할 수있다.\
하지만 FP로 분류된 캐릭터는 군집화가 잘 안되어있는 것을 볼수 있다.

```python
train_yhat = np.squeeze(model.predict(train_X))
train_yhat[train_yhat <= 0.5] = 0
train_yhat[train_yhat > 0.5] = 1
train_yhat = train_yhat.astype('int')
train_y_yhat = np.vstack([train_y, train_yhat]).T

test_yhat = np.squeeze(model.predict(test_X))
test_yhat[test_yhat <= 0.5] = 0
test_yhat[test_yhat > 0.5] = 1
test_yhat = test_yhat.astype('int')
test_y_yhat = np.vstack([test_y, test_yhat]).T

train_embeddings = layers.GlobalAveragePooling1D()(transformer_block(embedding_layer(train_X), training=False)).numpy()
test_embeddings = layers.GlobalAveragePooling1D()(transformer_block(embedding_layer(test_X), training=False)).numpy()

def make_label(y, yhat):
    if y == 0 and yhat == 0:
        return 'TN'
    elif y == 1 and yhat == 0:
        return 'FN'
    elif y == 0 and yhat == 1:
        return 'FP'
    else:
        return 'TP'
    
train_characters = pd.DataFrame(np.concatenate((train_embeddings, train_y_yhat), axis=1), columns=[i for i in range(train_embeddings.shape[1])] + ['y', 'yhat'])
train_characters['confusion'] = train_characters[['y', 'yhat']].apply(lambda row: make_label(row[0], row[1]), axis=1)

train_transformed = TSNE(n_components=2, verbose=1, perplexity=40, n_iter=300).fit_transform(train_characters[[0, 1, 2, 3, 4, 5, 6, 7]].values)
sns.scatterplot(x=train_transformed[:, 0], y=train_transformed[:, 1], hue=train_characters['confusion'].values, alpha=0.5)
```
